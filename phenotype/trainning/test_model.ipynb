```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Load the trained model
loaded_model = tf.keras.models.load_model("../models/1")

# Define class names (replace with your actual class names)
class_names = ['level1', 'level2', 'level3']

# Function to make predictions
# Function to make predictions
def predict_single_image(model, image_path, class_names):
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(256, 256))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Create a batch

    # Make predictions
    predictions = model.predict(img_array)
    predicted_class_index = np.argmax(predictions[0])
    predicted_class = class_names[predicted_class_index]
    
    # Convert raw predictions to percentages using softmax
    softmax_predictions = tf.nn.softmax(predictions[0])
    confidence_percentage = round(100 * (softmax_predictions[predicted_class_index].numpy()), 2)
    
    # Display the image and predictions
    plt.imshow(img)
    plt.title(f"Predicted Class: {predicted_class}\n Confidence: {softmax_predictions}%")
    plt.show()


```

```python
image_path = "./leaves.jpg"  # Replace with the path to your image
predict_single_image(loaded_model, image_path, class_names)
```

```python
import os
# Function to make predictions for all images in a folder
def predict_images_in_folder(model, folder_path, class_names):
    # Get a list of all files in the folder
    image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    
    # Iterate over each image file
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)

        # Load and predict for each image
        img = tf.keras.preprocessing.image.load_img(image_path, target_size=(256, 256))
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = tf.expand_dims(img_array, 0)  # Create a batch

        # Make predictions
        predictions = model.predict(img_array)
        predicted_class = class_names[np.argmax(predictions[0])]
        confidence = round(100 * (np.max(predictions[0])), 2)

        # Display the image and predictions
        plt.imshow(img)
        plt.title(f"Image: {image_file}\nPredicted Class: {predicted_class}\nConfidence: {confidence}%")
        plt.show()

# Example usage for a folder containing images
folder_path = "./tomatoes/day10"  # Replace with the path to your folder
predict_images_in_folder(loaded_model, folder_path, class_names)
```
